```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
from os.path import split, splitext, join
import shutil
from tqdm import tqdm
```

```python
data_path = '/Users/aleksandr/Desktop/hacks/health_data_hack/data/train'
```

```python
files = listdir(data_path)
```

```python
imgs, masks = [], []
pairs = {}
for file in files:
    name = splitext(file)[0]
    if name.endswith('_mask'):
        masks.append(file)
    else:
        imgs.append(file)
        pairs[file] = ''
```

```python
cnt = 0
for mask in masks:
    name = ''.join(mask.split('_mask'))
    if name in pairs.keys():
        pairs[name] = mask
        cnt += 1
```

```python
cnt
```

```python
img_path = '/Users/aleksandr/Desktop/hacks/health_data_hack/data/train_splited/images'
mask_path = '/Users/aleksandr/Desktop/hacks/health_data_hack/data/train_splited/masks'
for img, mask in tqdm(pairs.items()):
    if mask == '':
        continue
    shutil.copy(join(data_path, img), join(img_path, img))
    mask_name = ''.join(mask.split('_mask'))
    shutil.copy(join(data_path, mask), join(mask_path, mask_name))
```
